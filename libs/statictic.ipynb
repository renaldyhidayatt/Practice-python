```python
import statistics
 
 
data = [
    59.19,
    72.05,
    66.82,
    81.15,
    66.33,
    94.87,
    99.65,
    70.13,
    55.75,
    87.71,
    95.43,
    93.17,
    98.54,
    68.31,
    59.24,
    88.94,
    79.44,
    83.91,
    84.4,
    68.21,
]
 
print(round(statistics.mean(data), 2))
print(round(statistics.geometric_mean(data), 2))
print(round(statistics.harmonic_mean(data), 2))
print(round(statistics.median(data), 2))
```

```python

import random
 
 
data = [
    59.19,
    72.05,
    66.82,
    81.15,
    66.33,
    94.87,
    99.65,
    70.13,
    55.75,
    87.71,
    95.43,
    93.17,
    98.54,
    68.31,
    59.24,
    88.94,
    79.44,
    83.91,
    84.4,
    68.21,
]
 
random.seed(42)
mu = statistics.mean(data)
sigma = statistics.stdev(data)
dist = statistics.NormalDist(mu, sigma)
result = dist.samples(10)
result = [round(val, 3) for val in result]
print(result)
```
